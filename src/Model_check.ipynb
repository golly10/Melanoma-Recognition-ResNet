```python
from AuxFunctions import AuxFunctions
from ResNet import ResNet
from keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
```

```python
def predictImages(images, label):
    for i in range(0, len(images)):
        image = images[i].reshape(1, images[i].shape[0], images[i].shape[1], images[i].shape[2])
        Y_pred = model.predict(image)
        y_pred = np.argmax(Y_pred, axis=1)
        print("Real label -> ", label[i])
        print("Predicted label -> ", y_pred)
        plt.imshow(images[i])
        plt.show()
```

```python
model = ResNet.buildModel(0.00001, 1)
model.load_weights('logs/best_weights.hdf5')
data_malignant = []
label_malignant = np.zeros(10)
data_benign = []
label_benign = np.ones(10)
for i in range(1,10):
    data_malignant.append(np.array(image.load_img('../data/images-predict/malignant' + str(i) + '.jpg', target_size=[512,512])))
    data_benign.append(np.array(image.load_img('../data/images-predict/benign' + str(i) + '.jpg', target_size=[512,512])))
                       
print("-"*40)
print("Malignant images prediction")
predictImages(data_malignant, label_malignant)
print("-"*40)

print("-"*40)
print("Benign images prediction")
predictImages(data_benign, label_benign)
print("-"*40)
```
