```python
def readNoJson(data, masks):
    for r, d, f in os.walk("/Users/juanjo/code/Melanoma-Recognition-ResNet/data/images2"):
        for file in tqdm(f):
            paths = os.path.join(r, file)
            if paths[-5:] != 'Store' and paths[-1:] != 't' and paths[-5:] != 'n.png' and paths[-5:] != '.json':
                data.append(image.load_img(paths, target_size=[512,512]))
            if paths[-5:] != 'Store' and paths[-1:] != 't' and paths[-5:] == 'n.png':
                masks.append(image.load_img(paths, target_size=[512,512]))
    return data, masks
```
